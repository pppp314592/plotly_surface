```julia
using Plots 
plotly()
```

```julia
a=1
b=0.2
u=range(0,7π,length=51)
v=range(0.001,π/4,length=15)

x=@. a*cos(u)*sin(v')
y=@. a*sin(u)*sin(v')
z=@. a*(cos(v')+log(tan(v'/2)))+b*u;
```

```julia
surface(x, y, z,color=:lightrainbow,legend=false)
plot!(x,y,z,)
plot!(x',y',z')
```

```julia
gr()
plot(x,y,z,aspect_ratio=1,color=:black,legend=false,linewidth=0.3)
plot!(x',y',z',color=:black,linewidth=0.3)
```
